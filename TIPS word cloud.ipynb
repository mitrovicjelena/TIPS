```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("/Users/jelenamitrovic/Documents/IAA NCSU/Summer II/TIPS Assignment/WORD CLOUD/MSA22 TIPS Spreadsheet - Programming Tips.csv")
```

```python
df.head()
```

```python
df = df.drop(columns=['programming'])
```

```python
df.head()
```

## PROGRAMMING

```python
list_1 = df['2016'].dropna()
for i in range(2017, 2022):
    list_1 = list_1.append(df[str(i)].dropna())
df2 = pd.DataFrame()
df2['programming'] = list_1
df2
```

```python
#tip1 combined
programming = " ".join(tip for tip in df2.programming)
print ("There are {} words in the combination of all tips.".format(len(programming)))
```

```python
#generate the image
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(programming)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
wordcloud.to_file("Programming.png")
```
